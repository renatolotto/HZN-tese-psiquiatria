```python
import requests
import pandas as pd
from time import sleep

df = pd.read_excel('clinicas1.xls')
cne = df['CIDADE+CNES'].to_list()




headers = {'Accept': 'application/json, text/plain, */*',
'Accept-Encoding': 'gzip, deflate, br',
'Accept-Language': 'pt-BR,pt;q=0.9,en-US;q=0.8,en;q=0.7',
'Connection': 'keep-alive',
'Host': 'cnes.datasus.gov.br',
'Referer': 'https://cnes.datasus.gov.br/pages/estabelecimentos/consulta.jsp',
'sec-ch-ua': '" Not A;Brand";v="99", "Chromium";v="99", "Google Chrome";v="99"',
'sec-ch-ua-mobile': '?0',
'sec-ch-ua-platform': '"Windows"',
'Sec-Fetch-Dest': 'empty',
'Sec-Fetch-Mode': 'cors',
'Sec-Fetch-Site': 'same-origin',
'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}
id = []
cnes = []
noFantasia = []
noEmpresarial = []
natJuridica = []
natJuridicaMant = []
cnpj = []
tpPessoa = []
nvDependencia = []
nuAlvara = []
dtExpAlvara = []
orgExpAlvara = []
dsTpUnidade = []
dsStpUnidade = []
noLogradouro = []
nuEndereco = []
cep = []
regionalSaude = []
bairro = []
noComplemento = []
municipio = []
noMunicipio = []
uf = []
tpGestao = []
nuTelefone = []
tpSempreAberto = []
coMotivoDesab = []
dsMotivoDesab = []
cpfDiretorCln = []
stContratoFormalizado = []
nuCompDesab = []
dtCarga = []
dtAtualizacaoOrigem = []
dtAtualizacao = []

for cad in cne:
    try:
        print(cad)
        response = requests.get('https://cnes.datasus.gov.br/services/estabelecimentos/'+str(cad), headers=headers)
        print(response)
        print(response.text)
        c = response.json()
        id.append(c['id'])
        cnes.append(c['cnes'])
        noFantasia.append(c['noFantasia'])
        noEmpresarial.append(c['noEmpresarial'])
        natJuridica.append(c['natJuridica'])
        natJuridicaMant.append(c['natJuridicaMant'])
        cnpj.append(c['cnpj'])
        tpPessoa.append(c['tpPessoa'])
        nvDependencia.append(c['nvDependencia'])
        nuAlvara.append(c['nuAlvara'])
        dtExpAlvara.append(c['dtExpAlvara'])
        orgExpAlvara.append(c['orgExpAlvara'])
        dsTpUnidade.append(c['dsTpUnidade'])
        dsStpUnidade.append(c['dsStpUnidade'])
        noLogradouro.append(c['noLogradouro'])
        nuEndereco.append(c['nuEndereco'])
        cep.append(c['cep'])
        regionalSaude.append(c['regionalSaude'])
        bairro.append(c['bairro'])
        noComplemento.append(c['noComplemento'])
        municipio.append(c['municipio'])
        noMunicipio.append(c['noMunicipio'])
        uf.append(c['uf'])
        tpGestao.append(c['tpGestao'])
        nuTelefone.append(c['nuTelefone'])
        tpSempreAberto.append(c['tpSempreAberto'])
        coMotivoDesab.append(c['coMotivoDesab'])
        dsMotivoDesab.append(c['dsMotivoDesab'])
        cpfDiretorCln.append(c['cpfDiretorCln'])
        stContratoFormalizado.append(c['stContratoFormalizado'])
        nuCompDesab.append(c['nuCompDesab'])
        dtCarga.append(c['dtCarga'])
        dtAtualizacaoOrigem.append(c['dtAtualizacaoOrigem'])
        dtAtualizacao.append(c['dtAtualizacao'])
        sleep (0.5)
    except:
        pass

emp = pd.DataFrame({'id' : id
,'cnes' : cnes
,'Nome Fantasia' : noFantasia
,'Razão Social' : noEmpresarial
,'Natureza Jurídica' : natJuridica
,'Natureza Jurídica Mantenedora' : natJuridicaMant
,'CNPJ' : cnpj
,'Tipo de Pessoa' : tpPessoa
,'Nível de dependência' : nvDependencia
,'Número do Alvará' : nuAlvara
,'Data de expedição do Alvará' : dtExpAlvara
,'Orgão expedidor Alvará' : orgExpAlvara
,'Tipo de Unidade' : dsTpUnidade
,'Subtipo de Unidade' : dsStpUnidade
,'Logradouro' : noLogradouro
,'Número do endereço' : nuEndereco
,'CEP' : cep
,'Regional de Saude' : regionalSaude
,'Bairro' : bairro
,'Complemento' : noComplemento
,'Cod. Município' : municipio
,'Município' : noMunicipio
,'UF' : uf
,'Tipo de Gestão' : tpGestao
,'Telefone' : nuTelefone
,'Sempre aberto?' : tpSempreAberto
,'cód Motivo Desabilitação' : coMotivoDesab
,'Motitivo desabilitação' : dsMotivoDesab
,'CPF Diretor' : cpfDiretorCln
,'Situação contrato formalizado' : stContratoFormalizado
,'Número de Desabilitação' : nuCompDesab
,'Data da Carga' : dtCarga
,'Data da origem da atualização' : dtAtualizacaoOrigem
,'Data da Atualização' : dtAtualizacao})
```

```python
emp['CNPJ MANTENEDORA'] = None
emp['NOME MANTENEDORA'] = None

headers = {'Accept': 'application/json, text/plain, */*',
'Accept-Encoding': 'gzip, deflate, br',
'Accept-Language': 'pt-BR,pt;q=0.9,en-US;q=0.8,en;q=0.7',
'Connection': 'keep-alive',
'Host': 'cnes.datasus.gov.br',
'Referer': 'https://cnes.datasus.gov.br/pages/estabelecimentos/consulta.jsp',
'sec-ch-ua': '" Not A;Brand";v="99", "Chromium";v="99", "Google Chrome";v="99"',
'sec-ch-ua-mobile': '?0',
'sec-ch-ua-platform': '"Windows"',
'Sec-Fetch-Dest': 'empty',
'Sec-Fetch-Mode': 'cors',
'Sec-Fetch-Site': 'same-origin',
'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/99.0.4844.51 Safari/537.36'}
for i, rows in emp.iterrows():
    
    response = requests.get('https://cnes.datasus.gov.br/services/estabelecimentos-mantenedora/'+rows['Cod. Município']+rows['cnes'], headers=headers)
    print(response)
    print(response.text)
    try:
        mant = response.json()
        emp.loc[i,'CNPJ MANTENEDORA'] = mant['cnpjMantenedora']
        emp.loc[i,'NOME MANTENEDORA'] = mant['rzSocial']
    except:
        emp.loc[i,'CNPJ MANTENEDORA'] = None
        emp.loc[i,'NOME MANTENEDORA'] = None
```

```python
emp['leitos_saude_mental'] = None
emp['leitos_saude_mental_SUS'] = None
emp['leitos_hospital_dia'] = None
emp['leitos_hospital_dia_SUS'] = None
emp['leitos_psiquiatria'] = None
emp['leitos_psiquiatria_SUS'] = None
emp['leitos_neurologia'] = None
emp['leitos_nerologia_SUS'] = None
emp['leitos_geriatria'] = None
emp['leitos_geriatria_SUS'] = None
emp['leitos_geriatria_dia'] = None
emp['leitos_geriatria_dia_SUS'] = None
emp['leitos_total'] = None
emp['leitos_total_SUS'] = None

for i, rows in emp.iterrows():
    response = requests.get('https://cnes.datasus.gov.br/services/estabelecimentos-hospitalar/'+rows['Cod. Município']+rows['cnes'], headers=headers)
    print(response)
    print(response.text)
    serv = response.json()
    total = 0
    total_sus = 0
    mental = 0
    mental_sus = 0
    dia = 0
    dia_sus = 0
    psiq = 0 
    psiq_sus = 0
    neuro = 0
    neuro_sus = 0
    geri = 0
    geri_sus = 0
    geri_dia = 0
    geri_dia_sus = 0
    for s in serv:
        if s['coLeito'] == '87':
            mental = mental + int(s['qtExistente']) 
            mental_sus = mental_sus + int(s['qtSus'])
        if s['coLeito'] == '73':
            dia = dia + int(s['qtExistente']) 
            dia_sus = dia_sus + int(s['qtSus'])
        if s['coLeito'] == '47':
            psiq = psiq + int(s['qtExistente']) 
            psiq_sus = psiq_sus + int(s['qtSus'])
        if s['coLeito'] == '42':
            neuro = neuro + int(s['qtExistente']) 
            neuro_sus = neuro_sus + int(s['qtSus'])
        if s['coLeito'] == '72':
            geri = geri + int(s['qtExistente']) 
            geri_sus = geri_sus + int(s['qtSus'])
        if s['coLeito'] == '36':
            geri_dia = geri_dia + int(s['qtExistente']) 
            geri_dia_sus = geri_dia_sus + int(s['qtSus'])
        total = total + int(s['qtExistente'])
        total_sus = total_sus + int(s['qtSus'])
    emp.loc[i,'leitos_saude_mental'] = mental
    emp.loc[i,'leitos_saude_mental_SUS'] = mental_sus
    emp.loc[i,'leitos_total'] = total
    emp.loc[i,'leitos_total_SUS'] = total_sus
    emp.loc[i,'leitos_hospital_dia'] = dia
    emp.loc[i,'leitos_hospital_dia_SUS'] = dia_sus
    emp.loc[i,'leitos_psiquiatria'] = psiq
    emp.loc[i,'leitos_psiquiatria_SUS'] = psiq_sus
    emp.loc[i,'leitos_neurologia'] = neuro
    emp.loc[i,'leitos_nerologia_SUS'] = neuro_sus
    emp.loc[i,'leitos_geriatria'] = geri
    emp.loc[i,'leitos_geriatria_SUS'] = geri_sus
    emp.loc[i,'leitos_geriatria_dia'] = geri_dia
    emp.loc[i,'leitos_geriatria_dia_SUS'] = geri_dia_sus
```

```python
emp['MEDICO NEUROLOGISTA'] = None
emp['MEDICO PSIQUIATRA'] = None
emp['MEDICO NEUROFISIOLOGISTA CLINICO'] = None
emp['PSICOLOGO CLINICO'] = None
emp['PSICOLOGO EDUCACIONAL'] = None
emp['PSICOLOGO HOSPITALAR'] = None
emp['PSICOLOGO SOCIAL'] = None
emp['PSICOLOGO DO TRABALHO'] = None
emp['NEUROPSICOLOGO'] = None
emp['PSICANALISTA '] = None
emp['PSICOLOGO ACUPUNTURISTA'] = None
emp['TOTAL FUNCIONARIOS'] = None
for i, rows in emp.iterrows():
    response = requests.get('https://cnes.datasus.gov.br/services/estabelecimentos-profissionais/'+rows['Cod. Município']+rows['cnes'], headers=headers)
    print(response)
    print(response.text)
    serv = response.json()
    total = 0
    mn = 0
    mp = 0
    mnc = 0
    pc = 0
    pe = 0
    ph = 0
    ps = 0
    pt = 0
    np = 0
    pca = 0
    pa = 0
    for s in serv:
        if s['cbo'] == '225112':
            mn += 1
        if s['cbo'] == '225133':
            mp += 1
        if s['cbo'] == '225350':
            mnc += 1
        if s['cbo'] == '251505':
            pc += 1
        if s['cbo'] == '251510':
            pe += 1
        if s['cbo'] == '251520':
            ph += 1
        if s['cbo'] == '251530':
            ps += 1
        if s['cbo'] == '251540':
            pt += 1
        if s['cbo'] == '251545':
            np += 1
        if s['cbo'] == '251550':
            pca += 1
        if s['cbo'] == '251555':
            pa += 1
        total += 1
    emp.loc[i,'MEDICO NEUROLOGISTA'] = mn
    emp.loc[i,'MEDICO PSIQUIATRA'] = mp
    emp.loc[i,'MEDICO NEUROFISIOLOGISTA CLINICO'] = mnc
    emp.loc[i,'PSICOLOGO CLINICO'] = pc
    emp.loc[i,'PSICOLOGO EDUCACIONAL'] = pe
    emp.loc[i,'PSICOLOGO HOSPITALAR'] = ph
    emp.loc[i,'PSICOLOGO SOCIAL'] = ps
    emp.loc[i,'PSICOLOGO DO TRABALHO'] = pt
    emp.loc[i,'NEUROPSICOLOGO'] = np
    emp.loc[i,'PSICANALISTA '] = pca
    emp.loc[i,'PSICOLOGO ACUPUNTURISTA'] = pa
    emp.loc[i,'TOTAL FUNCIONARIOS'] = total
```
